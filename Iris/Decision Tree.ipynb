## Decision Tree
Training and Visualizing a Decision Tree
### Importing the libraries

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

### Preparing the data

```python
iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Classification model

```python
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X_train, y_train)
```

### Visulaize the trained model

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(tree_clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names = iris.feature_names[2:],
                class_names=iris.target_names)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Iris.png')
Image(graph.create_png())
```

This command line converts the .dot file to a .png image file `dot -Tpng iris_tree.dot -o iris_tree.png` you can convert to any format. One of the many qualities of Decision Trees is that they require
very little data preparation. In particular, they don’t require feature
scaling or centering at all.

* Desicion trees: white box models
* Random Forests, neural networks: black box models

Decision Trees Params to reduce the risk of overfitting recommended
* max_depth 
* min_samples_leaf

### Regression model

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(max_depth=2)
tree_reg.fit(X_train, y_train)
```

```python
dot_data = StringIO()
export_graphviz(tree_reg, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names = iris.feature_names[2:],
                class_names=iris.target_names)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Iris.png')
Image(graph.create_png())
```
